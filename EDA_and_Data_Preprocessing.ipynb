```python
#Importing the libraries - pandas
import pandas as pd
```

## Movie Box Office Dataset

```python
# Loading the data into Pandas DataFrames
movie_bo = pd.read_csv("Data/Movie Box Office.csv")
movie_bo
```

```python
movie_bo.info()
```

```python
movie_bo.isnull().sum()
# Number of null values in each column:
```

```python
# Dropping the rows with null values
movie_bo.dropna(subset = ['ReleaseDate'], inplace = True)
movie_bo.isnull().sum()
# Number of null values in each column:
```

```python
movie_bo.duplicated(subset=['Movie', 'ReleaseDate', 'Rank']).sum()
# Number of repeated columns (Movies):
```

```python
# Removing the repeated rows:
movie_bo.drop_duplicates(subset=['Movie', 'ReleaseDate', 'Rank'], inplace=True)
movie_bo.duplicated(subset=['Movie', 'ReleaseDate', 'Rank']).sum()
# Number of repeated columns (Movies):
```

```python
# Creating a unique ID and saving the data
movie_bo.to_csv('Data/Movie Box Office.csv')
x = pd.read_csv('Data/Movie Box Office.csv')
x.rename(columns={'Unnamed: 0': 'ID'}, inplace=True)
x.to_csv('Data/Movie Box Office.csv', index=False)
```

## Streaming Titles Dataset

```python
# Loading the data into Pandas DataFrames
streaming_df = pd.read_csv("Data/streaming_titles.csv", encoding='latin1')
streaming_df
```

From the above, seems like some data got wrapped into the following rows. Let's clean it by removing null values first...

```python
streaming_df.info()
```

```python
streaming_df.isnull().sum()
# Number of null values in each column:
```

```python
# Dropping the rows with null values
streaming_df.dropna(subset = ['type', 'title', 'date_added'], inplace = True)
streaming_df.isnull().sum()
# Number of null values in each column:
```

```python
# Checking for the wrapped data
streaming_df
```

```python
set(streaming_df['type'])
```

The wrapped data is now cleaned and only Movies and TV Shows remain.

```python
streaming_df.duplicated(subset=['type', 'title', 'date_added']).sum()
# Number of repeated columns (Movies):
```
